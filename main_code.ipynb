```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
X_train=pd.read_csv('X_train.csv', engine='python')
X_test=pd.read_csv('X_test.csv', engine='python')
y_train=pd.read_csv('y_train.csv', engine='python')
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
X_train.loc[:,['주구매상품', '주구매지점']]=X_train.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
X_train.loc[:,['주구매상품','주구매지점']].head()
```

```python
X_test.loc[:,['주구매상품','주구매지점']]=X_test.loc[:,['주구매상품','주구매지점']].apply(LabelEncoder().fit_transform)
X_test.loc[:,['주구매상품','주구매지점']].head()
```

```python
X_train.loc[:,['환불금액']]=X_train.loc[:,['환불금액']].fillna(0)
X_train.loc[:,['환불금액']].head()
```

```python
X_test.loc[:,['환불금액']]=X_test.loc[:,['환불금액']].fillna(0)
X_test.loc[:,['환불금액']].head()
```

```python
X_test_head=X_test.iloc[:,0]
X_test_head
```

```python
X_train=X_train.iloc[:,1:]
X_test=X_test.iloc[:,1:]
y_train=y_train.iloc[:,-1]
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

## Logistic Regression

```python
LR=LogisticRegression()
LR.fit(X_train, y_train)
print('LR score:', LR.score(X_train, y_train))
predict=LR.predict_proba(X_test)
predict=pd.DataFrame(predict)
predict=predict.iloc[:,1]
predict=pd.concat([X_test_head,predict], axis=1)
print(predict)
```

## MLPClassifier

```python
mlp=MLPClassifier(hidden_layer_sizes=(10,), activation='relu', solver='adam', alpha=0.0001, batch_size='auto', learning_rate_init=0.001)
mlp.fit(X_train,y_train)
print('MLP score:', mlp.score(X_train,y_train))
predict=mlp.predict_proba(X_test)
predict=pd.DataFrame(predict)
predict=predict.iloc[:,1]
predict=pd.concat([X_test_head, predict])
print(predict)
```

## RandomForestClaasifier

```python
RF=RandomForestClassifier(n_estimators=100, max_depth=5, random_state=999)
RF.fit(X_train, y_train)
print('RF score :', RF.score(X_train,y_train))
predict=RF.predict_proba(X_test)
predict=pd.DataFrame(predict)
predict=predict.iloc[:,1]
predict=pd.concat([X_test_head, predict], axis=1)
print(predict)
```
